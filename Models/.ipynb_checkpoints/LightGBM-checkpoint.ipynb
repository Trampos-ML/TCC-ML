```python
import pandas as pd
import lightgbm as lgb
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder

# Carregar o DataFrame com os dados dos filmes
df = pd.read_csv('dados.csv')

# Vetorizar as sinopses
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['Sinopse'])

# Inicializar o codificador de rótulos
label_encoder = LabelEncoder()

# Converter os rótulos para valores numéricos
df['Label'] = label_encoder.fit_transform(df['Titulo'])

# Definir o dataset de treinamento com os rótulos numéricos
train_data = lgb.Dataset(X, label=df['Label'])

# Definir os parâmetros do modelo LightGBM
params = {
    'objective': 'multiclass',
    'num_class': len(df['Titulo'].unique()),
    'random_state': 42
}

# Treinar o modelo
model_lgb = lgb.train(params, train_data)

# Fazer a previsão usando o modelo
entrada_usuario = input("Digite suas preferências: ")
entrada_usuario_vector = vectorizer.transform([entrada_usuario])
filme_recomendado_lgb = model_lgb.predict(entrada_usuario_vector).argmax()

# Converter o índice predito de volta para o título do filme
filme_recomendado_titulo = label_encoder.inverse_transform([filme_recomendado_lgb])[0]

print("Filme recomendado para você (LightGBM):", filme_recomendado_titulo)
```
