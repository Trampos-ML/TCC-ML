```python
import pandas as pd
from surprise import Reader, Dataset, KNNBasic
from surprise.model_selection import train_test_split, cross_validate
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score
import numpy as np

# Carregar o dataset
dados = pd.read_csv('dados.csv')

# Inicializar o leitor
reader = Reader(rating_scale=(1, 5))

# Carregar o dataset no formato correto para o Surprise
data = Dataset.load_from_df(dados[['userID', 'movieID', 'rating']], reader)

# Dividir o dataset em treino e teste
trainset, testset = train_test_split(data, test_size=0.2)

# Inicializar o algoritmo KNN
algo = KNNBasic()

# Treinar o modelo
algo.fit(trainset)

# Executar validação cruzada
cv_results = cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)

# Preparar dados para avaliação de classificação
testset = trainset.build_testset()
predictions = algo.test(testset)
true_ratings = [pred.r_ui for pred in predictions]
predicted_ratings = [pred.est for pred in predictions]
predicted_labels = np.round(predicted_ratings)  # Arredondar para obter a classificação binária

# Calcular métricas de classificação
accuracy = accuracy_score(true_ratings, predicted_labels)
precision = precision_score(true_ratings, predicted_labels)
conf_matrix = confusion_matrix(true_ratings, predicted_labels)

# Mostrar os resultados da validação cruzada
print("Resultados da validação cruzada:")
for measure in ['test_rmse', 'test_mae']:
    print(f"{measure}: {cv_results[measure].mean()}")

# Mostrar as métricas de classificação
print("Métricas de classificação:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Confusion Matrix:")
print(conf_matrix)
```
