# Importação das dependências e arquivos csv

```python
import pandas as pd

# Caminho para o arquivo CSV
caminho_arquivo = 'Data/netflix_titles.csv'

# Carregar o arquivo CSV em um dataframe
df = pd.read_csv('Data/netflix_titles.csv')


# Exibir o dataframe
print(df)
```

# Checagem dos dados e suas respectivas colunas

```python
df.info()
```

# Remoção das colunas que não serão usadas

```python
df = df.drop(columns=['show_id', 'type','cast','date_added','rating','listed_in'])
df
```

# Remoção dos seriados pela coluna duration

```python
df = df[~df['duration'].apply(lambda x: str(x).lower().find('season') != -1)]
print(df)
```
