```python
# Lista para armazenar os textos extraídos
extracted_texts = []

text_files = [
    '/content/drive/My Drive/Limpeza de Texto/Dengue-1.txt',
    '/content/drive/My Drive/Limpeza de Texto/Dengue-2.txt',
    '/content/drive/My Drive/Limpeza de Texto/Dengue-3.txt',
    '/content/drive/My Drive/Limpeza de Texto/Dengue-4.txt',
    '/content/drive/My Drive/Limpeza de Texto/Dengue-5.txt',
    '/content/drive/My Drive/Limpeza de Texto/Dengue-6.txt',
    '/content/drive/My Drive/Limpeza de Texto/Dengue-7.txt',
    '/content/drive/My Drive/Limpeza de Texto/Dengue-8.txt',
    '/content/drive/My Drive/Limpeza de Texto/Dengue-9.txt',
    '/content/drive/My Drive/Limpeza de Texto/Dengue-10.txt',
    # Adicione mais caminhos de arquivos, se necessário
]

# Definir os padrões regex e os textos
regex_patterns = [
    'Dengue.*?monitoring\.',
    'Dengue is an important.*?vaccine development\.',
    'Dengue is a mosquito.*?prevention\.',
    'Dengue is.*?of dengue\.',
    'Changing dengue epidemiological.*?needed\.',
    'Dengue is an arboviral.*?discussed\.',
    'Dengue fever.*?fever\.',
    'Dengue is a highly.*?individuals\.',
    'Dengue virus.*?everywhere\.',
    'Since the.*?significance\.',
    # Adicione mais padrões regex, se necessário
]

# Iterar sobre os padrões regex e os textos
for regex_pattern in regex_patterns:
    print("Usando o padrão regex:", regex_pattern)
    for text_file in text_files:
        # Ler o conteúdo do arquivo
        with open(text_file, 'r') as f:
            full_text = f.read()

        # Use expressão regular para encontrar o trecho desejado
        match = re.search(regex_pattern, full_text, re.DOTALL)

      # Verifique se houve uma correspondência antes de armazenar
        if match:
            extracted_text = match.group(0)
            print(extracted_text, '\n')  # Adicionando uma quebra de linha após cada impressão
            # Adicionar o texto extraído à lista
            extracted_texts.append(extracted_text)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

# Baixar recursos necessários do NLTK
nltk.download('wordnet')

# Inicializar o lematizador
lemmatizer = WordNetLemmatizer()

# Função para realizar o tratamento de texto
def preprocess_text(text):
    # Converter o texto para minúsculas
    text = text.lower()
    # Remover caracteres especiais e pontuações
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenização: dividir o texto em palavras
    tokens = word_tokenize(text)
    # Remover stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    # Lematização: converter as palavras para sua forma base
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
    return lemmatized_tokens

# Aplicar o tratamento de texto aos resumos extraídos
preprocessed_texts = [preprocess_text(text) for text in extracted_texts]

# Imprimir os primeiros 10 tokens de cada resumo após o tratamento de texto
for i, tokens in enumerate(preprocessed_texts):
    print(f"Resumo {i+1}: {tokens[:10]}")
```

```python
from nltk import pos_tag

# Função para buscar sujeitos e adjetivos em um texto
def find_subjects_and_adjectives(text):
    # Realizar POS tagging para marcar as classes gramaticais das palavras
    tagged_words = pos_tag(text)

    # Inicializar listas para armazenar sujeitos e adjetivos
    subjects = []
    adjectives = []

    # Iterar sobre as palavras marcadas e identificar sujeitos e adjetivos
    for word, tag in tagged_words:
        if tag.startswith('NN'):  # NN: substantivo
            subjects.append(word)
        elif tag.startswith('JJ'):  # JJ: adjetivo
            adjectives.append(word)

    return subjects, adjectives

# Buscar sujeitos e adjetivos em cada resumo
for i, text in enumerate(preprocessed_texts):
    subjects, adjectives = find_subjects_and_adjectives(text)
    print(f"Resumo {i+1}:")
    print("Sujeitos:", subjects)
    print("Adjetivos:", adjectives)
    print()
```

```python
# Inicializar o spaCy
nlp = spacy.load('en_core_web_sm')

# Função para realizar análise avançada de texto
def analyze_text(text):
    # Realizar análise de texto com spaCy
    doc = nlp(" ".join(text))

    # Extrair constituintes (frases, cláusulas, etc.)
    constituents = [chunk.text for chunk in doc.noun_chunks]

    # Extrair entidades nomeadas
    named_entities = [(entity.text, entity.label_) for entity in doc.ents]

    return constituents, named_entities

# Função para buscar elementos morfológicos específicos em um texto
def find_morphological_elements(text):
    # Realizar POS tagging para marcar as classes gramaticais das palavras
    tagged_words = pos_tag(text)

    # Inicializar listas para armazenar diferentes tipos de elementos morfológicos
    nouns = []
    verbs = []
    adjectives = []
    adverbs = []
    # Adicione mais listas para outros tipos de elementos, se necessário

    # Iterar sobre as palavras marcadas e identificar diferentes tipos de elementos morfológicos
    for word, tag in tagged_words:
        if tag.startswith('NN'):  # NN: substantivo
            nouns.append(word)
        elif tag.startswith('VB'):  # VB: verbo
            verbs.append(word)
        elif tag.startswith('JJ'):  # JJ: adjetivo
            adjectives.append(word)
        elif tag.startswith('RB'):  # RB: advérbio
            adverbs.append(word)
        # Adicione mais condições para outros tipos de elementos, se necessário

    return nouns, verbs, adjectives, adverbs
```

```python
# Analisar cada resumo
for i, text in enumerate(preprocessed_texts):
    # Realizar análise avançada de texto
    constituents, named_entities = analyze_text(text)

    # Buscar elementos morfológicos
    nouns, verbs, adjectives, adverbs = find_morphological_elements(text)

    # Imprimir os resultados
    print(f"Resumo {i+1}:")
    print("Constituintes:", constituents)
    print("Entidades nomeadas:", named_entities)
    print("Substantivos:", nouns)
    print("Verbos:", verbs)
    print("Adjetivos:", adjectives)
    print("Advérbios:", adverbs)
    print()
```

```python
# Converter os textos pre-processados de volta para strings
preprocessed_texts_str = [" ".join(text) for text in preprocessed_texts]

# Criar o vetorizador TF-IDF
vectorizer = TfidfVectorizer()

# Aplicar o vetorizador nos textos
tfidf_matrix = vectorizer.fit_transform(preprocessed_texts_str)

# Obter as palavras do vocabulário
vocabulary = vectorizer.get_feature_names_out()

# Calcular a frequência de cada palavra nos documentos
word_frequencies = tfidf_matrix.sum(axis=0)

# Converter as frequências em uma lista ordenada de tuplas (palavra, frequência)
word_freq_tuples = [(word, word_frequencies[0, idx]) for word, idx in vectorizer.vocabulary_.items()]

# Ordenar as palavras por frequência
word_freq_tuples_sorted = sorted(word_freq_tuples, key=lambda x: x[1], reverse=True)

# Imprimir as palavras mais frequentes e suas frequências
print("Palavras mais frequentes:")
for word, freq in word_freq_tuples_sorted[:10]:
    print(f"{word}: {freq}")
```

```python
print("\nAnálise do resultado:")
print("As palavras mais frequentes nos resumos são geralmente termos que aparecem com maior frequência em todos os documentos. Essas palavras podem não ser muito informativas sobre o conteúdo específico dos resumos.")
print("Por outro lado, as palavras mais relevantes são aquelas que têm valores TF-IDF mais altos. Isso significa que essas palavras são distintas e aparecem com mais frequência em documentos específicos, tornando-as mais importantes para entender o conteúdo dos resumos.")
print("Portanto, ao analisar os resumos usando TF-IDF, devemos prestar atenção não apenas nas palavras mais frequentes, mas também nas palavras mais relevantes, que podem fornecer insights valiosos sobre o conteúdo dos documentos.")
```

```python
from gensim import corpora, models

# Criar um dicionário a partir dos textos pré-processados
dictionary = corpora.Dictionary(preprocessed_texts)

# Criar um corpus (bag of words) a partir do dicionário
corpus = [dictionary.doc2bow(text) for text in preprocessed_texts]

# Definir o número de tópicos
num_topics = 5

# Treinar o modelo LDA
lda_model = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)

# Exibir os tópicos aprendidos pelo modelo
for idx, topic in lda_model.print_topics(-1):
    print(f"Tópico {idx}: {topic}")
```
